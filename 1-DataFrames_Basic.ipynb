#### Drop Rows with Null Values:

```python
nba = pd.read_csv("files/nba.csv")
nba.tail(3)
```

```python
# we have a lot of null(NaN) value in data source. We can drop these values with .dropna() function. 

nba.dropna()
```

```python
# it moves row if all columns are null. if we add "inplace=True" in parenthesis,it will be moved permanently. 

nba.dropna(how = "all", inplace=True)
```

```python
# axis = 0 rows, axis = 1 columns

nba.dropna(axis = 1)
# or
nba.dropna(axis = "columns")
```

```python
# Only if those null values exist in the Salary column. 
nba.dropna(subset =["Salary"]) 
```

```python
nba.dropna(subset =["Salary","College"]) 
```

#### Fill in Null Values with the .fillna() Method:

```python
nba = pd.read_csv("files/nba.csv")
nba.head(3)
```

```python
# to fill with zero instead null value: 

nba.fillna(0)
```

```python
#to fill in specific column's value: 

nba["Salary"].fillna(0, inplace = True)
nba.head(3)
```

```python
nba["College"].fillna("No College",inplace = True)
nba.head(6)
```

#### The .astype() Method:

```python
nba = pd.read_csv("files/nba.csv").dropna(how = "all")
nba["Salary"].fillna(0,inplace =True)
nba["College"].fillna("None", inplace =True)
nba.head()
```

```python
nba.dtypes
nba.info()
```

```python
nba["Salary"] = nba["Salary"].astype("int")
nba.head(3)
```

```python
nba["Number"] = nba["Number"].astype("int")
nba.head(3)
```

```python
nba["Age"] = nba["Age"].astype("int")
nba.head(3)
```

```python
nba["Age"].astype("float")
```

```python
nba["Position"].nunique()
```

```python
nba["Position"] = nba["Position"].astype("category")
```

```python
nba["Team"] = nba["Team"].astype("category")
```

```python
nba.head()
```

```python
nba.info()
```

#### Sort a DataFrame with the .sort_values() Method, Part I:

```python
nba = pd.read_csv("files/nba.csv")
```

```python
nba.sort_values("Name", ascending = False)
nba.head(3)
```

```python
nba.sort_values("Age", ascending = False)
nba.head()
```

```python
nba.sort_values("Salary", ascending = False, inplace = True)
nba.head(3)
```

```python
# na_position specify null value's positons in value. 

nba.sort_values("Salary", ascending = False, na_position = "first").tail()
```

#### Sort a DataFrame with the .sort_values() Method, Part II:

```python
nba = pd.read_csv("files/nba.csv")
```

```python
# it sort columns by alphabetically. Firstly sort by team and then sort by name. 

nba.sort_values(["Team","Name"]).head()
```

```python
# it sorts columns reverse 

nba.sort_values(["Team","Name"],ascending = False).head()
```

```python
# if we want to sort columns by different way we can use this method:  

nba.sort_values(["Team","Name"],ascending = [True,False]).head()
```

#### Sort DataFrame with the .sort_index() Method:

```python
nba = pd.read_csv("files/nba.csv")
```

```python
nba.sort_values(["Number","Salary","Name"], inplace = True)
nba.tail(3)
```

```python
# to turn back original shape. 
nba.sort_index()
```

```python
# reverse index orders
nba.sort_index(ascending = False, inplace = True)
nba.head()
```

#### Rank Values with the .rank() Method:

```python
nba = pd.read_csv("files/nba.csv").dropna(how="all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by = "Salary", ascending = False)
```
